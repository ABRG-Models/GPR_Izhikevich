#From the Izhikevich ABC formulation to the I-V form

## The ABC form
Suppose you have an Izhikevich Simple Model given in the form:

$$ \dot{v} = Av^2 + Bv + C - u + I \tag{1} $$
$$ \dot{u} = a(bv - u) \tag{2} $$

where if $v > v_{peak}$ then $v \leftarrow c; u \leftarrow u + d$

$v$ is the membrane voltage, $u$ is a recovery variable, $A$, $B$ and $C$, $a$, $b$, $c$ and $d$ are dimensionless model parameters and $v_{peak}$ is the reset voltage.

You may be starting from one of the canonical models given in Izhikevich's 2003 IEEE paper, with a plan to tweak the parameters. Perhaps you like the thalamo-cortical neuron for which:

$$A=0.04; B=5; C=140; a=0.02; b=0.25; c=-65; d=0.05; v_{peak}=30$$

(See Fig. 2, TC from Izhikevich, "Simple Model of Spiking Neurons", IEEE Transactions on Neural Networks, Vol 14, 2003.)

So that:

$$ \dot{v} = 0.04v^2 + 5v + 140 - u + I $$
$$ \dot{u} = 0.02(0.25v - u) $$
$$ v > 30: v \leftarrow -65; u \leftarrow u + 0.05 $$

## Graphing the ABC form
Let's graph the vector field and the nullclines for this system. First up, we'll import the matplotlib and numpy libraries, which are used in the rest of this notebook.

```python
# Setup first
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Now we'll set up model parameters and some parameters used to select ranges over which we'll compute the nullclines and vector field of the system.

```python
# Model parameters
A=0.04;
B=5;
C=140;
a=0.02;
b=0.25;
c=-65;
d=0.05;
vpeak=30;
I=0; # constant input current

# Graphing parameters
lowerV = -120;
upperV = 30;
qvlowerv = lowerV;
qvupperv = upperV;
qvloweru = -20;
qvupperu = 40;
qpscale = 3;
qpdivisor = .00001;
```

Having set up the parameters, lets compute nullclines and the vector field. The vector field is simply the solution to Equations (1) and (2). The nullclines are the lines at which the components of the vector fields are zero. The v nullcline is found by setting Equation (1) to zero and rearranging to give $u$ in terms of $v$:

$$\dot{v} = Av^2 + Bv + C - u + I = 0$$
$$\implies u = Av^2 + Bv + C + I$$

Similarly for the u nullcline:

$$\dot{u} = a(bv - u) = 0$$
$$\implies bv - u = 0$$
$$\implies u = bv$$

The same thing in python looks like:

```python
# Compute nullclines
Vn = np.linspace (lowerV, upperV, 1000);
nc_v = A*Vn*Vn + B*Vn + C + I;
nc_u = b * Vn;
        
# Compute vector field
thescale = abs(qvlowerv-qvupperv)/qpscale;
Vgv = np.linspace(qvlowerv, qvupperv, thescale);
ugv = np.linspace(qvloweru, qvupperu, thescale);
Vg, ug = np.meshgrid(Vgv, ugv);
vdot = A*Vg*Vg + B*Vg + C - ug + I;
udot = a*(b*Vg - ug);
```

We can now graph nc_v, nc_u, vdot and udot in the $v/u$ phase plane:

```python
plt.figure(figsize=(18,10))
plt.clf;

Q = plt.quiver(Vg, ug, vdot, udot*10, pivot='mid', color='k', width=0.001, scale=2000)
lef,rgt,bot,top = plt.axis()
plt.axis([lef, rgt, bot, top])

plt.plot(Vn, nc_v, color='b')
plt.plot(Vn, nc_u, color='r')

plt.title('Izhikevich Simple Model: ABC formulation; TC model')
plt.xlabel('v');
plt.ylabel('u');
```

## Converting to the I-V form
Now we get to the nub of this article. Suppose you'd prefer to work with the Simple Model expressed in the I-V form:

$$ C_a \dot{v} = k(v-v_r)(v-v_t) - u' + I' \tag{3} $$
$$ \dot{u'} = a'(b'(v-v_r) - u') \tag{4} $$

$$ v > v_{peak}: v \leftarrow c'; u' \leftarrow u' + d' $$

Where here, $C_a$ is the capacitance, $k$ is ?, $v_r$ is the resting potential, $v_t$ is a threshold potential and $a'$, $b'$, $c'$ and $d'$ are dimensionless model parameters as before. The recovery variable u is expressed as $u'$, in case a change of variables should prove necessary (it will!). Similar for the current, $I'$. We'll assume that $v$ does not require any scaling as ultimately we want the IV model to reproduce the same function $v(t)$ as the ABC model.

How can we translate from the "ABC" formulation given by Izhikevich into this form, which he gives in his 2007 book? Can we express Equations (1) and (2) in this form to produce the identical phase plane diagram above?


### The quadratic nullcline
First, lets expand Eq. (3) and compare it with Eq. (1):

$$C_a \dot{v} = k(v^2 - v v_r - v v_t + v_r v_t) - u' + I' \tag{5}$$

$$\implies C_a \dot{v} = kv^2 - k(v_r + v_t)v + k v_r v_t - u' + I'$$

$$\implies \dot{v} = \frac{k}{C_a}v^2 - \frac{k(v_r + v_t)}{C_a}v + \frac{k v_r v_t}{C_a} - \frac{u'}{C_a} + \frac{I'}{C_a}$$

compare with Eq (1):

$$\dot{v} = Av^2 + Bv + C - u + I \tag{1}$$

You can see already that we'll need to make a change of variable in $u$, to deal with the factor $\frac{1}{C_a}$. Any injected or synaptic currents will also need to be scaled by $\frac{1}{C_a}$.

Let  $u = \frac{u'}{C_a}$ and $I = \frac{I'}{C_a}$ so that Eq. (5) becomes:

$$ \dot{v} = \frac{k}{C_a}v^2 - \frac{k(v_r + v_t)}{C_a}v + \frac{k v_r v_t}{C_a} - u + I \tag{6}$$

From Eqs. (1) and (6) we can write:

$$ A = \frac{k}{C_a} \qquad B = - \frac{k(v_r + v_t)}{C_a} \qquad C = \frac{k v_r v_t}{C_a} \tag{7}$$

Now lets find $v_r$ and $v_t$ in terms of $A$, $B$ and $C$. From (7):

$$C = A v_r v_t$$
$$\implies v_t = \frac{C}{A v_r}$$
$$B = -A(v_r + v_t)$$
$$\implies B = -A(v_r + \frac{C}{A v_r}) = -A v_r - \frac{C}{v_r}$$
$$\implies B v_r = -A v_r^2 - C$$
$$\implies A {v_r}^2 + B v_r + C = 0$$

So, $v_r$ is simply one of the roots of the quadratic in Eq. (1)! We could have carried out the same procedure to obtain an identical expression for $v_t$ in terms of $A$, $B$ and $C$ indicating that $v_r$ is one root of the quadratic and $v_t$ is the other. This is not surprising from the form of Eq. (3).

Using the quadratic formula, we find $v_r$, $v_t$:

$$v_r, v_t = {-B\pm\sqrt{B^2-4 A C}\over 2 A}; A=0.04; B=5; C=140 \implies v_r\approx-82.66, v_t\approx-42.34 $$

Now what about $k$ and $C_a$? In fact, we're free to choose these so long as $\frac{k}{C_a}=A=0.04$. In the model, these would be given biologically realistic values, but here I'll set $C_a = 1$ so that $k=A$.

###The linear nullcline
Let's look at the equations for $\dot{u}$.

$$ \dot{u} = a(bv - u) \tag{2} $$
and
$$ \dot{u'} = a'(b'(v-v_r) - u') \tag{4} $$

We already found that $u={u'\over C_a}$ so by applying the chain rule for differentiation:

$$ \dot{u'} =  {du'\over dt} = {du'\over du}{du \over dt} = C_a {du\over dt} = C_a \dot{u} $$

we can then write Eq. (4) like this:

$$ C_a \dot{u} = a'(b'(v-v_r) - C_a u) $$

Now consider the nullcline for this system; the set of solutions to:

$$ C_a \dot{u} = a'(b'(v-v_r) - C_a u) = 0 $$
$$\implies b'(v-v_r) - C_a u = 0$$
$$\implies b'(v-v_r) = C_a u$$ 
$$\implies u = { b'\over C_a}(v-v_r) = { b'\over C_a} v - { b'\over C_a} v_r \tag{8}$$

Now compare with the nullcline of Eq. 2:

$$ \dot{u} = a(bv - u) = 0 $$
$$ \implies u = bv$$

These are *not* the same lines. $v_r$ is in general non-zero, so the two systems must be different. If it weren't for $v_r$, then we could set $b={b'\over C_a}$ and be done. Let's graph both systems, letting $b={b'\over C_a} = {b'\over 1}$ and $a=a'$; $c=c'$; $d=d'$.

## Graphing Try 1 of the IV formulation

```python
# Model parameters
Ca=1;
k=0.04;
vr = -82.66;
vt = -42.34;
a=0.02;
b=0.25;
c=-65;
d=0.05;
vpeak=30;
I=0; # constant input current

# Graphing parameters
lowerV = -120;
upperV = 30;
qvlowerv = lowerV;
qvupperv = upperV;
qvloweru = -20;
qvupperu = 40;
qpscale = 3;
qpdivisor = .00001;
```

From Eq. (6) we have the quadratic nullcline:
$$ \dot{v} = \frac{k}{C_a}v^2 - \frac{k(v_r + v_t)}{C_a}v + \frac{k v_r v_t}{C_a} - u + I = 0$$
$$ \implies u = \frac{k}{C_a}v^2 - \frac{k(v_r + v_t)}{C_a}v + \frac{k v_r v_t}{C_a}  + I$$
$$ \implies u = {k\over C_a}(v-v_r)(v-v_t) + I $$

Eq. (8) gives the linear nullcline.

```python
# Compute nullclines
Vn = np.linspace (lowerV, upperV, 1000);
nc_v = (k/Ca)*(Vn-vr)*(Vn-vt) + I;
nc_u = b * (Vn-vr);
        
# Compute vector field
thescale = abs(qvlowerv-qvupperv)/qpscale;
Vgv = np.linspace(qvlowerv, qvupperv, thescale);
ugv = np.linspace(qvloweru, qvupperu, thescale);
Vg, ug = np.meshgrid(Vgv, ugv);
vdot = (k/Ca)*(Vg-vr)*(Vg-vt) - ug + I;
udot = a*(Ca*b*(Vg-vr) - Ca*ug);
```

```python
plt.figure(figsize=(18,10))
plt.clf;

Q = plt.quiver(Vg, ug, vdot, udot*10, pivot='mid', color='k', width=0.001, scale=2000)
lef,rgt,bot,top = plt.axis()
plt.axis([lef, rgt, bot, top])

plt.plot(Vn, nc_v, color='b')
plt.plot(Vn, nc_u, color='r')

plt.title('Izhikevich Simple Model: IV formulation Try 1; TC model')
plt.xlabel('v');
plt.ylabel('u');
```

Comparing the graph above with the ABC formulation graph, you'll see that the quadratic nullcline looks the same (note that the scale on the graph has not changed) but the linear nullcline is translated up, compared with the linear nullcline in the ABC formulation graph. Here, the $u$ nullcline *always* passes through $u=0$ at $v=v_r$. In the ABC formulation, the $u$ nullcline passes through $u=0$ at $v=0$. This means that our "Try 1" at unifying the two systems is done with; we need to do something else.

## Try 2

To obtain the identical system, we need to translate the linear nullcline. We can do this by adding an offset to the $u$ variable.

### Linear nullcline

We already have $u={u'\over C_a} \implies u' = C_a u$. Now add an offset to let

$$u' = C_a u + u_0$$

Now Eq. (4) becomes:

$$ \dot{u'} = a'(b'(v-v_r) - u') \tag{4} $$
$$\implies C_a \dot{u} = a'(b'(v-v_r) - (C_a u + u_0)) = 0 $$
$$\implies b'(v-v_r) - C_a u - u_0 = 0$$
$$\implies b'(v-v_r) - u_0 = C_a u$$ 
$$\implies u = {b'\over C_a}(v-v_r) - {u_0\over C_a} \tag{9}$$

We can find $u_0$ by requiring that $u(v)=0$ for $v=0$:

$$ 0={b'\over C_a}(0-v_r) - {u_0\over C_a}$$
$$ \implies -{b'\over C_a}v_r = {u_0\over C_a}$$
$$ \implies u_0 = -b' v_r \tag{10}$$

This gives us our final expression for the $\dot{u}$ vector field component.

$$C_a \dot{u} = a'(b'(v-v_r) - (C_a u - b' v_r)) \tag{11}$$

### Quadratic nullcline

We're not quite done. Because $u$ appears in the equation for $\dot{v}$, we have to re-compute v_r and v_t. Starting from Eq. (3) again:

$$ C_a \dot{v} = k(v-v_r)(v-v_t) - u' + I' \tag{3} $$

and substitute in the expression for $u'$:

$$ C_a \dot{v} = k(v-v_r)(v-v_t) - (C_a u - b'v_r) + I' $$

We now expand this out and compare with Eq. (1):

$$ C_a \dot{v} = k(v^2 -v_r v - v_t v + v_r v_t) - (C_a u - b'v_r) + I' $$
$$ \implies \dot{v} = {k\over C_a}(v^2 -v_r v - v_t v + v_r v_t) - u + {b'\over C_a}v_r + {I'\over C_a} $$
$$ \implies \dot{v} = {k\over C_a}v^2 - {k(v_r + v_t)\over C_a} v + {k v_r v_t\over C_a}  + {b'\over C_a}v_r - u  + {I'\over C_a} $$

We can now compare with Eq. (1) and write:

$$ A = \frac{k}{C_a} \qquad B = - \frac{k(v_r + v_t)}{C_a} \qquad C = \frac{k (v_t + b') v_r}{C_a} \tag{12}$$

$$\implies B = -A(v_r+v_t) \qquad C=A v_t v_r$$
$$\implies v_r = -{B\over A} - v_t \qquad v_r = {C\over A(v_t + b')}$$
$$\implies -{B\over A} - v_t = {C\over A(v_t + b')}$$
$$\implies -{B\over A}A (v_t+b') - v_t A (v_t+b') = C$$
$$\implies -B v_t - B b' - A {v_t}^2 - A b' v_t = C$$
$$\implies A {v_t}^2 + (B+A b')v_t + (C+B b') = 0$$
$$\implies 0.04 {v_t}^2 + 5.01 v_t + 141.25 = 0$$

We now have our quadratic to solve (expressed in $v_t$ this time, rather than $v_r$). Solution using the quadratic formula gives $v_r$, $v_t$ as:
$$ v_r \approx -82.38963066 \qquad v_t \approx -42.86036934 $$

As before, our choice of C_a and k have been constrained only by the requirement that ${k\over C_a} = 0.04$. We chose $k=0.04$, $C_a=1$. This gives our final equation for $\dot{v}$, which we write down here along with the expression for $\dot{u}$

$$ \dot{v} = 0.04(v-v_r)(v-v_t) - (u - b v_r) + I \tag{13} $$
$$ \dot{u} = a(b(v-v_r) - (u - b v_r)) \tag{14}$$

Let's graph these expressions to see if they're equivalent to Eqs. (1) and (2).

```python
# Model parameters
Ca=1;
k=0.04;
vr = -82.38963066;
vt = -42.86036934;
a=0.02;
b=0.25;
c=-65;
d=0.05;
vpeak=30;
I=0; # constant input current

# Graphing parameters
lowerV = -120;
upperV = 30;
qvlowerv = lowerV;
qvupperv = upperV;
qvloweru = -20;
qvupperu = 40;
qpscale = 3;
qpdivisor = .00001;
```

```python
# Compute nullclines
Vn = np.linspace (lowerV, upperV, 1000);
nc_v = (k/Ca)*(Vn-vr)*(Vn-vt) + b*vr + I;
nc_u = b * Vn;
        
# Compute vector field
thescale = abs(qvlowerv-qvupperv)/qpscale;
Vgv = np.linspace(qvlowerv, qvupperv, thescale);
ugv = np.linspace(qvloweru, qvupperu, thescale);
Vg, ug = np.meshgrid(Vgv, ugv);
vdot = (k/Ca)*(Vg-vr)*(Vg-vt) - ug + b*vr + I;
udot = a*(Ca*b*(Vg-vr) - Ca*ug + b*vr);
```

```python
plt.figure(figsize=(18,10))
plt.clf;

Q = plt.quiver(Vg, ug, vdot, udot*10, pivot='mid', color='k', width=0.001, scale=2000)
lef,rgt,bot,top = plt.axis()
plt.axis([lef, rgt, bot, top])

plt.plot(Vn, nc_v, color='b')
plt.plot(Vn, nc_u, color='r')

plt.title('Izhikevich Simple Model: IV formulation Try 2; TC model')
plt.xlabel('v');
plt.ylabel('u');
```
